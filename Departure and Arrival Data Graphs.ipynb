```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
total_data = pd.read_csv("Resources/cleaned_df.csv", low_memory=False)
airline_df = pd.read_csv("Resources/airlines.csv", low_memory=False)
airport_df = pd.read_csv("Resources/airports.csv", low_memory=False)

#total_data['ORIGIN_AIRPORT']
outbound = total_data.loc[total_data["ORIGIN_AIRPORT"]=="Minneapolis-Saint Paul International Airport",:]
inbound= total_data.loc[total_data["DEST_AIRPORT"]=="Minneapolis-Saint Paul International Airport",:]
```

```python
carrier_df = outbound[['OP_UNIQUE_CARRIER','DEP_DELAY','ARR_DELAY']]
unique_df = carrier_df['OP_UNIQUE_CARRIER'].unique()
listing = unique_df.tolist()
listing
departure_df = carrier_df['DEP_DELAY']
data = {'index': listing, 'Departure Delay Mean':[0,0,0,0,0,0,0,0,0], 'Departure Delay Mode':[0,0,0,0,0,0,0,0,0], 'Departure Delay Median':[0,0,0,0,0,0,0,0,0], 'Arrival Delay Mean':[0,0,0,0,0,0,0,0,0],'Arrival Delay Median':[0,0,0,0,0,0,0,0,0],'Arrival Delay Mode':[0,0,0,0,0,0,0,0,0]}
bar_plot_df = pd.DataFrame(data=data)

bar_plot_df

```

```python
departure_mean_bar = bar_plot_df['Departure Delay Mean'].tolist()
departure_delay_mode = bar_plot_df['Departure Delay Mode'].tolist()
departure_delay_median = bar_plot_df['Departure Delay Median'].tolist()


X = ["DELTA", "SPIRIT", "SOUTHWEST", "ALASKAN", "UNITED", "AMERICAN", "FRONTIER", "ENDEAVOR", "SKYWEST"]
Y = departure_mean_bar
Z = departure_delay_mode
R= departure_delay_median

df = pd.DataFrame(np.c_[Y,Z,R], index=X )
ax = df.plot(kind='bar', fontsize=20, width=.75)
ax.legend(["MEAN", "MEDIAN", "MODE"]);
ax.set_xlabel("Airlines")
ax.set_ylabel("Minutes")
ax.set_title("Departure Metrics")

fig = ax.get_figure()
fig.savefig("Resources/departure_bar_chart.png")
```
